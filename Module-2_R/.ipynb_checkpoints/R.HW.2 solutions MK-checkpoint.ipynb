## 3. Prime numbers
### Find first 500 prime numbers. What is the 500th prime number?

```r
prime500 <- c()
x <- 1
while (length(prime500) < 500) {
    x <- x + 1
    #if (x == 2) {
     #   FALSE
    #} else if (any(x %% 2:(x-1) == 0)){
    #    FALSE
    #} else {
    #    print(x)
    #    prime500 <- c(prime500,x)
    #}
    if(x==2){
        prime500 <- c(prime500, x)
    }
    else if(!any(x %% 2:(x-1) == 0)){
        prime500 <- c(prime500, x)
    }
}
prime500[500] ##found this command online that gives me the 500th value in a vector
prime500[1:100]
length(prime500)
```

## 4. Fibonacci number
### Define a function to return the Xth fibonacci number
https://en.wikipedia.org/wiki/Fibonacci_number

```r
Fibbi <- function(y) {
n1 = 0
n2 = 1
count = 2
    if(y < 0) {
        print("No negatives please")
        } else if (y == 0) {
            print(paste(y,"-th","Fibonacci number:"))
            print(n1)
    } else {
        if(y == 1) {
            print(paste(y,"-th","Fibonacci number:"))
            print(n1)
        } else {
            
            #print(n1)
            #print(n2)
            while (count < y) {
                nth = n1 + n2
                #print(nth)
                # update values
                n1 = n2
                n2 = nth
                count = count + 1
                    
            }
            print(paste(y,"-th","Fibonacci number:", nth))
        }
    }
}

Fibbi(9)
```

### 5.4 Define a function to evaluate padj output:
    - padj > 0.05: nonsig
    - padj <= 0.05: sig

```r
# padj
# padj > 0.05: nonsig
# padj <= 0.05: sig

padjType <- function(padj) {
    if (is.na(padj)) {
        return(NA)
    }
    else if(padj > 0.05) {
        return("nonsig")
    }
    else if(padj <= 0.05) {
        return("sig")
    }
}
```

### 5.5 Apply the function to both nav.mem and nav.d14 and create one column to add the evaluation result.

```r
nav.d14.padjType <- lapply(df.nav.d14$padj, padjType)
nav.d14.padjType <- factor(unlist(nav.d14.padjType))
nav.d14.2 <- cbind(df.nav.d14, nav.d14.padjType )
head(nav.d14.2)
```

```r
nav.mem.padjType <- lapply(df.nav.mem$padj, padjType)
nav.mem.padjType <- factor(unlist(nav.mem.padjType))
nav.mem.2 <- cbind(df.nav.mem, nav.mem.padjType )
head(nav.mem.2)
```
