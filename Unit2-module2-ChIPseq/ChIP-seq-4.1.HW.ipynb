3.For all the alignment output sam files:

3.1 convert to bam, sort and filter

```python
%%bash
cd /home/mkissai/Applied-Bioinformatics-HW-local/Sample_fastq/Sample_fastq_trimmed
source activate chip
for i in *.fq.sam
do
samtools view -bS "$i" > "$i".bam
done

for j in $(seq -f %1.0f 3001750 3001776)
do
mv SRR$j.500_trimmed.fq.sam.bam SRR$j.500_trimmed.fq.bam
done
```

```python
%%bash
cd /home/mkissai/Applied-Bioinformatics-HW-local/Sample_fastq/Sample_fastq_trimmed
source activate chip
for i in *.bam
do
samtools sort "$i" -o "$i".srt.bam
done

for j in $(seq -f %1.0f 3001750 3001776)
do
mv SRR$j.500_trimmed.fq.bam.srt.bam SRR$j.500_trimmed.fq.srt.bam
done
```

```python
%%bash
cd /home/mkissai/Applied-Bioinformatics-HW-local/Sample_fastq/Sample_fastq_trimmed
source activate chip
for i in *.srt.bam
do
samtools rmdup -S "$i" "$i".srt.dupr.bam
done

for j in $(seq -f %1.0f 3001750 3001776)
do
mv SRR$j.500_trimmed.fq.srt.bam.srt.dupr.bam SRR$j.500_trimmed.fq.srt.dupr.bam
done
```

3.2 Summarize the sizes for all the sam files / bam files / sorted bam files / sorted and filtered bam files into a txt file named “sum_sizes.txt”

```python
%%bash
cd /home/mkissai/Applied-Bioinformatics-HW-local/Sample_fastq/Sample_fastq_trimmed

for j in $(seq -f %1.0f 3001750 3001776)
do
stat SRR$j.500_trimmed.fq* >> sum_stat_all.txt
done

awk 'NR%8==1,NR%8==2' sum_stat_all.txt > sum_sizes.txt
cat sum_sizes.txt
```
